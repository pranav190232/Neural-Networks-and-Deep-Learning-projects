```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2
import numpy as np
import math

uploaded = files.upload()

image_path = next(iter(uploaded))
img_pil = Image.open(image_path)
img = np.array(img_pil)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)  

model = YOLO("yolov8s.pt")  

classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"]

object_counts = {"car": 0, "bus": 0, "truck": 0, "motorbike": 0}

results = model(img, stream=True,conf=0.15, iou=0.25)

for r in results:
    boxes = r.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        conf = math.ceil((box.conf[0] * 100)) / 100

        cls = int(box.cls[0])
        currentclass = classNames[cls]

        if currentclass in ['car', 'bus', 'truck', 'motorbike'] and conf > 0.15:
            object_counts[currentclass] += 1

            cv2.rectangle(img, (x1, y1), (x2, y2), (255, 0, 255), 1)

            cv2.putText(img, f'{currentclass} {conf:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.25, (255, 255, 255), 1)


img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.axis('off')
plt.show()

print(f"Final counts: {object_counts}")
```

```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2
import numpy as np
import math

uploaded = files.upload()

image_path = next(iter(uploaded))
img_pil = Image.open(image_path)
img = np.array(img_pil)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) 

model = YOLO("yolov8s.pt")

classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"]

object_counts = {"car": 0, "bus": 0, "truck": 0, "motorbike": 0}

results = model(img, stream=True, conf=0.15, iou=0.25)

img_with_labels = img.copy()
img_with_boxes_only = img.copy()

for r in results:
    boxes = r.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        conf = math.ceil((box.conf[0] * 100)) / 100

        cls = int(box.cls[0])
        currentclass = classNames[cls]

        if currentclass in ['car', 'bus', 'truck', 'motorbike'] and conf > 0.15:
            object_counts[currentclass] += 1

            cv2.rectangle(img_with_labels, (x1, y1), (x2, y2), (255, 0, 255), 1)
            cv2.rectangle(img_with_boxes_only, (x1, y1), (x2, y2), (255, 0, 255), 1)

            cv2.putText(img_with_labels, f'{currentclass} {conf:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.25, (255, 255, 255), 1)

img_with_labels_rgb = cv2.cvtColor(img_with_labels, cv2.COLOR_BGR2RGB)
img_with_boxes_only_rgb = cv2.cvtColor(img_with_boxes_only, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_with_labels_rgb)
plt.title('Detected Objects with Labels')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(img_with_boxes_only_rgb)
plt.title('Detected Objects with Bounding Boxes')
plt.axis('off')

plt.show()

print(f"Final counts: {object_counts}")
```

```python
from google.colab import files
from PIL import Image
import matplotlib.pyplot as plt
from ultralytics import YOLO
import cv2
import numpy as np
import math

uploaded = files.upload()

image_path = next(iter(uploaded))
img_pil = Image.open(image_path)
img = np.array(img_pil)
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)

model = YOLO("yolov8s.pt")

classNames = ["person", "bicycle", "car", "motorbike", "aeroplane", "bus", "train", "truck", "boat",
              "traffic light", "fire hydrant", "stop sign", "parking meter", "bench", "bird", "cat",
              "dog", "horse", "sheep", "cow", "elephant", "bear", "zebra", "giraffe", "backpack", "umbrella",
              "handbag", "tie", "suitcase", "frisbee", "skis", "snowboard", "sports ball", "kite", "baseball bat",
              "baseball glove", "skateboard", "surfboard", "tennis racket", "bottle", "wine glass", "cup",
              "fork", "knife", "spoon", "bowl", "banana", "apple", "sandwich", "orange", "broccoli",
              "carrot", "hot dog", "pizza", "donut", "cake", "chair", "sofa", "pottedplant", "bed",
              "diningtable", "toilet", "tvmonitor", "laptop", "mouse", "remote", "keyboard", "cell phone",
              "microwave", "oven", "toaster", "sink", "refrigerator", "book", "clock", "vase", "scissors",
              "teddy bear", "hair drier", "toothbrush"]

object_counts = {"car": 0, "bus": 0, "truck": 0, "motorbike": 0}
tracker_id = 1

img_with_tracker_ids = img.copy()

results = model(img, stream=True, conf=0.15, iou=0.25)

for r in results:
    boxes = r.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0]
        x1, y1, x2, y2 = int(x1), int(y1), int(x2), int(y2)

        conf = math.ceil((box.conf[0] * 100)) / 100

        cls = int(box.cls[0])
        currentclass = classNames[cls]

        if currentclass in ['car', 'bus', 'truck', 'motorbike'] and conf > 0.15:
            object_counts[currentclass] += 1

            cv2.rectangle(img_with_tracker_ids, (x1, y1), (x2, y2), (255, 0, 255), 1)

            cv2.putText(img_with_tracker_ids, f'{tracker_id}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.25, (255, 255, 255), 1)
            tracker_id += 1

img_with_tracker_ids_rgb = cv2.cvtColor(img_with_tracker_ids, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 5))
plt.imshow(img_with_tracker_ids_rgb)
plt.title('Detected Objects with Tracker IDs')
plt.axis('off')
plt.show()

print(f"Final counts: {object_counts}")
```
